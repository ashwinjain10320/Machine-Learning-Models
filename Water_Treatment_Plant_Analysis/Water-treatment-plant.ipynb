```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
```

```python
index = ['Q-E','ZN-E','PH-E','DBO-E','DQO-E','SS-E','SSV-E','SED-E','COND-E','PH-P','DBO-P','SS-P','SSV-P','SED-P',
         'COND-P','PH-D','DBO-D','DQO-D','SS-D','SSV-D','SED-D','COND-D','PH-S','DBO-S','DQO-S','SS-S','SSV-S','SED-S',
         'COND-S','RD-DBO-P','RD-SS-P','RD-SED-P ','RD-DBO-S','RD-DQO-S','RD-DBO-G','RD-DQO-G','RD-SS-G','RD-SED-G'] 
```

```python
data = pd.read_table('water-treatment.data',sep=',',header=None,index_col=0)
```

```python
data.columns= index
```

```python
data
```

```python
d = data.copy()
```

```python
for i in d.columns:
    d[i] = np.where(d[i]=='?',np.nan,d[i])
```

```python
d.isnull().sum()
```

```python
data = d
```

```python
data.head()
```

```python
for i in data.columns:
    data[i] = data[i].astype('float64')
```

```python
data.ffill(axis=0,inplace=True)
```

```python
x = [ i for i in data.columns if (data[i].isnull().sum()>0)]
x
```

```python
for i in x:
    data[i].fillna(data[i].mean(),inplace=True)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca.fit(data)
```

```python
r_data = pca.transform(data)
r_data = pd.DataFrame(r_data,columns=['PCA1','PCA2'])
r_data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
n_clusters = [i for i in range(2,13)]
wcss = [ ]
for i in n_clusters:
    km = KMeans(n_clusters=i,init='k-means++')
    km.fit(r_data)
    wcss.append(km.inertia_)
```

```python
plt.plot(n_clusters,wcss)
```

```python
plt.scatter(r_data['PCA1'],r_data['PCA2'])
```

```python
km = KMeans(n_clusters=4,init='k-means++')
km.fit(r_data)
```

```python
r_data['predict'] = km.predict(r_data)
```

```python
plt.scatter(r_data['PCA1'],r_data['PCA2'],c=r_data['predict'])
```

```python
data['cluster'] = km.predict(r_data)
```
